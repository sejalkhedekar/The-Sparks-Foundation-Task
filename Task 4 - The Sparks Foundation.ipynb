## Preprocessing Data for Visualization Purpose

```python
# renaming column names
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
data.head()
```

```python
data.dropna(axis=1,how='all') #dropna() method allows the user to analyze and drop Rows/Columns with Null values in different ways.
```

```python
# finally these are the required columns
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude','Attacktype','kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

## Exploratory Data Analysis

### Data Insights

## 1.Most Affected Regions 

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data['Region'].value_counts()[:10].index,data['Region'].value_counts()[:10].values)
plt.title('Most Affected Regions in each Year')
plt.xlabel('Number of Attacks')
plt.ylabel('Count')
plt.xticks(rotation=90);
```

## 2.Top 10 Affected Counteries

```python
fig=plt.subplots(figsize=(15,7))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='BuPu')
plt.title('Top 10 Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

## 3. Most Number of People Killed (Country)

```python
fig=plt.figure(figsize = (10,5))
data.groupby(['Country'])['kill'].sum().sort_values(ascending = False).head(15).plot(kind = 'bar', color = 'green')

plt.xticks(rotation=90)
plt.ylabel("Killed")
plt.show()
```

## 4.Attack Type Used by Terrorist

```python
fig = plt.figure(figsize=(15,6))
sns.countplot('Attacktype',data=data,order=data['Attacktype'].value_counts().index,palette='Dark2')
plt.xticks(rotation=90)
plt.title(" Attack Type used by Terrorist")
```

## 5.Top 10 most active Terrorist organisations

```python
fig=data['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='red',figsize=(10,5))
plt.title("Top 10 most active Terrorist organisations",fontsize=20)
plt.xlabel("Terrorist organisations",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

## 6.Frequent target type of terrorist

```python
fig = plt.figure(figsize=(20,10))
sns.countplot('Targettype',data=data,order=data['Targettype'].value_counts().index,palette='Set1')
plt.xticks(rotation=90)
plt.title(" frequent target type of terrorist")
```

## 7.Most Attacked Cities

```python
data['City'].value_counts()
fig=plt.figure(figsize=(15,7))
sns.barplot(x=data['City'].value_counts()[:20].index,y=data['City'].value_counts()[:20].values, palette = "autumn_r")
plt.title('Most Attacked Cities')
plt.xticks(rotation=40)
plt.xlabel('City')
plt.ylabel('Attacks')
plt.show()
```

## 8.Terrorist Attacks Per Year

```python
fig=plt.subplots(figsize=(15,10))
sns.countplot(x='Year',data=data,palette='bone_r')
plt.xticks(rotation=60)
plt.title('Terrorist Attacks Per Year')
plt.ylabel('Attacks')
plt.xlabel('Year')
plt.show()
```
